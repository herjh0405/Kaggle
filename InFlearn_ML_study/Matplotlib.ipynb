## matplotlib

```python
import matplotlib.pyplot as plt

X=range(100)
Y=[value**2 for value in X]
plt.plot(X,Y)
plt.show()
```

```python
# graph  figure 
import numpy as np

X_1=range(100)
Y_1=[np.cos(value) for value in X]

X_2=range(100)
Y_2=[np.sin(value) for value in X]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
```

```python
fig=plt.figure() # figure 
fig.set_size_inches(10,5)
ax_1=fig.add_subplot(1,2,1)
ax_2=fig.add_subplot(1,2,2)

ax_1.plot(X_1, Y_1, c='b',linestyle='dashed',label=
         'label1')
ax_2.plot(X_2, Y_2, c='c',label='label2')

plt.grid(True,lw=0.4,ls='--',c='.90')
plt.xlim(-100,200)
plt.ylim(-100,200)

plt.legend(shadow=True, fancybox=True, loc='lower right')
plt.title('$y=\\frac{ax+b}{test}$')
plt.savefig('test.png',c='a')
plt.show()
```

## Bar Chart

```python
data = [[5., 25., 50., 20.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]]

X = np.arange(0,8,2)
```

```python
plt.bar(X + 0.00, data[0], color = 'b', width = 0.50)
plt.bar(X + 0.50, data[1], color = 'g', width = 0.50)
plt.bar(X + 1.0, data[2], color = 'r', width = 0.50)
plt.xticks(X+0.50, ("A","B","C", "D"))
plt.show()
```

## Boston house price

```python
import pandas as pd

data_url='https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data'
df_data=pd.read_csv(data_url,sep='\s+',header=None)
df_data.columns = [
    'CRIM','ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO' ,'B', 'LSTAT', 'MEDV'] 
# Column Header  
df_data.head()
```

```python
fig = plt.figure()
fig.set_size_inches(10.0, 10.0, forward=True)
ax = []
for i in range(1,5):
    ax.append(fig.add_subplot(2,2,i))

columns = ["CRIM", "PTRATIO", "AGE", "NOX"]
colors = ["b", "g", "c", "r"]
for i in range(4):
    ax[i].scatter(df_data[columns[i]], df_data["MEDV"], color=colors[i], label=columns[i])
    ax[i].legend()
    ax[i].set_title(columns[i])

plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

```python
df_data.plot()
plt.show()
```

```python
# histogram   
fig = plt.figure()
fig.set_size_inches(5,5)

ax_1 = fig.add_subplot(1,2,1)
ax_2 = fig.add_subplot(1,2,2)
ax_1.plot(df_data["MEDV"])
ax_2.hist(df_data["MEDV"], bins=50)
ax_1.set_title("House price MEDV")
ax_2.set_title("House price MEDV")

plt.show()
```

```python
#     scale  
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
scale_data = std_scaler.fit_transform(df_data)
scale_data
```

```python
# boxplot 0 outlier, - 
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.boxplot(scale_data, labels=df_data.columns)
fig.set_size_inches(10, 10)
plt.show()
```

```python
corr_data = np.corrcoef(scale_data.T)
corr_data
```

```python
corr_data.shape
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr_data, vmin=-1, vmax=1,  interpolation='nearest')
fig.colorbar(cax)
fig.set_size_inches(10,10)
ticks= np.arange(0,13,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)

ax.set_xticklabels(df_data.columns)
ax.set_yticklabels(df_data.columns)

plt.show()
```
