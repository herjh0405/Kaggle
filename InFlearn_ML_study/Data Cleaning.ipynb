## 결측치
* 데이터가 없으면 sample을 drop
* 데이터가 없는 최소 갯수를 정해서 sample을 drop (데이터의 갯수가 별로 없을 때)
* 데이터가 거의 없는 feature를 drop (feature=column)
* 최빈값, 평균값으로 데이터를 채움

```python
import pandas as pd
import numpy as np
```

```python
raw_data = {'first_name': ['Jason', np.nan, 'Tina', 'Jake', 'Amy'],
        'last_name': ['Miller', np.nan, 'Ali', 'Milner', 'Cooze'],
        'age': [42, np.nan, 36, 24, 73],
        'sex': ['m', np.nan, 'f', 'm', 'f'],
        'preTestScore': [4, np.nan, np.nan, 2, 3],
        'postTestScore': [25, np.nan, np.nan, 62, 70]}
df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'sex', 'preTestScore', 'postTestScore'])
df
```

```python
# null값 확인
df.isnull().sum()
df.isnull().sum() / len(df) # 몇퍼센트의 값이 null이냐
```

```python
# 가장 쉬운 방법, null이 있으면 삭제
df_no_missing=df.dropna()
df_no_missing
```

```python
df_cleaned=df.dropna(how='all') # 모든 데이터가 null이면 drop
df_cleaned
```

```python
df['location']=np.nan
df
```

```python
df.dropna(axis=1, thresh=4) # 값이 4개가 안되면 삭제
```

```python
df['preTestScore'].mean() # 평균값
df['preTestScore'].median() # 중위값
df['preTestScore'].mode() # 최빈값
```

```python
df.fillna(0) # null을 0으로 채움
```

```python
df['preTestScore'].fillna(df['preTestScore'].mean(), inplace=True)
df
```

```python
# 성별을 group으로 묶어서 걔네들의 평균값으로 바꿈
df["postTestScore"].fillna(
    df.groupby("sex")["postTestScore"].transform("mean"), inplace=True)
df
```

```python
df[df['age'].notnull() & df['sex'].notnull()]
```

## Data Binning
* 구간별로 나눠줌

```python
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'],
        'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'],
        'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'],
        'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}
df = pd.DataFrame(raw_data, columns = ['regiment', 'company', 'name', 'preTestScore', 'postTestScore'])
df
```

```python
bins=[0,25,50,75,100]
group_names=['Low','Okay','Good','Great']
categories=pd.cut(df['postTestScore'],bins,labels=group_names)
categories
```

```python
df['categories'] = pd.cut(df['postTestScore'], bins, labels=group_names)
pd.value_counts(df['categories'])
```

## Feature Scaling
* Min-Max Normalization
* Z-score Noramlization

### Min-Max Normalization

```python
df = pd.DataFrame({'A':[14.00,90.20,90.95,96.27,91.21],'B':[103.02,107.26,110.35,114.23,114.68], 'C':['big','small','big','small','small']})
df
```

```python
df["A"] = ( df["A"] - df["A"].min() )  \
/ (df["A"].max() - df["A"].min()) * (5 - 1) + 1
df
```

###  Z-score Noramlization
* 평균을 빼고 표준편차로 나눠줌 -1.96~1.96 상하위 95% 정도

```python
df["B"] = ( df["B"] - df["B"].mean() )  \
/ (df["B"].std() )
df
```

```python
df = pd.io.parsers.read_csv(
    'https://raw.githubusercontent.com/rasbt/pattern_classification/master/data/wine_data.csv',
     header=None,
     usecols=[0,1,2]
    )

df.columns=['Class label', 'Alcohol', 'Malic acid']

df
```

```python
df_std = std_scaler.transform(df[['Alcohol', 'Malic acid']])
df_std
```

```python
minmax_scaler = preprocessing.MinMaxScaler().fit(df[['Alcohol', 'Malic acid']])
minmax_scaler =minmax_scaler.transform(df[['Alcohol', 'Malic acid']])
minmax_scaler
```
